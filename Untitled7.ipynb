```python
# ================================================================
#   RELIANCE.csv — Complete BiLSTM Forecasting Pipeline
# ================================================================
#   Step 0  — Imports & Seed
#   Step 1  — Load Raw Data
#   Step 2  — Add Technical Indicators        (Concept 1)
#   Step 3  — Convert to Stationary Features  (Concept 2)
#   Step 4  — Sequence Creation               (Multi-Step y)
#   Step 5  — Model Builder                   (Bidirectional LSTM)
#   Step 6  — Walk-Forward Validation         (Concept 3)
#   Step 7  — Final Forecast on Unseen Data
#   Step 8  — Plot Everything
#   Step 9  — Save Everything
# ================================================================


# ================================================================
# STEP 0 — IMPORTS & SEED
# ================================================================
import os, random, warnings, joblib
import numpy             as np
import pandas            as pd
import matplotlib.pyplot as plt
import matplotlib.dates  as mdates
import matplotlib.gridspec as gridspec
warnings.filterwarnings("ignore")

from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics       import mean_absolute_error, mean_squared_error

import tensorflow as tf
from tensorflow.keras.models    import Sequential
from tensorflow.keras.layers    import LSTM, Bidirectional, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau

# ── Fix all randomness sources ───────────────────────────────
def set_seed(seed: int = 42) -> None:
    os.environ["PYTHONHASHSEED"]       = str(seed)
    os.environ["TF_DETERMINISTIC_OPS"] = "1"
    random.seed(seed)
    np.random.seed(seed)
    tf.random.set_seed(seed)

set_seed(42)

# ── Global config (change only these) ────────────────────────
CSV_PATH         = "RELIANCE.csv"
LOOKBACK         = 60      # input window — past N days
FORECAST_HORIZON = 10      # output window — predict N days ahead
BATCH_SIZE       = 32
EPOCHS           = 100
N_SPLITS         = 5       # walk-forward folds
TARGET           = "Target_Return"


# ================================================================
# STEP 1 — LOAD RAW DATA
# ================================================================
def load_data(path: str) -> pd.DataFrame:
    """
    Reads RELIANCE.csv.
    - Parses DD-MM-YYYY dates  (dayfirst=True)
    - Reorders to standard OHLCV
    - Replaces zero-volume rows with NaN and drops them
      (117 rows in RELIANCE have Volume=0 which causes log(0)=inf)
    """
    print("\n" + "="*65)
    print("  STEP 1 — Load Raw Data")
    print("="*65)

    df = pd.read_csv(path)
    df["Date"] = pd.to_datetime(df["Date"], dayfirst=True)
    df.set_index("Date", inplace=True)
    df.sort_index(inplace=True)                    # oldest → newest

    df = df[["Open", "High", "Low", "Close", "Volume"]]
    df = df.apply(pd.to_numeric, errors="coerce")

    # Zero-volume rows produce log(0) = -inf in Step 3 → fix here
    df["Volume"] = df["Volume"].replace(0, np.nan)
    df.dropna(inplace=True)

    print(f"  Rows loaded  : {len(df):,}")
    print(f"  Date range   : {df.index[0].date()}  →  {df.index[-1].date()}")
    print(f"  Last Close   : ₹{df['Close'].iloc[-1]:.2f}")
    print(f"  Columns      : {list(df.columns)}")
    return df


# ================================================================
# STEP 2 — ADD TECHNICAL INDICATORS   (Concept 1)
# ================================================================
#
#   WHY: Close alone only tells you "what" happened.
#        Indicators tell you "how" and "why":
#          MACD / MA → trend direction & momentum shifts
#          RSI        → overbought / oversold conditions
#          Bollinger  → volatility squeeze / breakout
#          OBV        → volume confirms (or diverges from) price
#          Stochastic → short-term momentum
#
#   RULE: Compute ALL indicators from RAW prices NOW.
#         Step 3 converts raw prices to returns AFTER this.
#
def add_technical_indicators(df: pd.DataFrame) -> pd.DataFrame:
    print("\n" + "="*65)
    print("  STEP 2 — Technical Indicators  (Concept 1)")
    print("="*65)

    d = df.copy()

    # ── TREND ───────────────────────────────────────────────────
    d["EMA_12"]      = d["Close"].ewm(span=12, adjust=False).mean()
    d["EMA_26"]      = d["Close"].ewm(span=26, adjust=False).mean()
    d["MACD"]        = d["EMA_12"] - d["EMA_26"]
    d["MACD_Signal"] = d["MACD"].ewm(span=9, adjust=False).mean()
    d["MACD_Hist"]   = d["MACD"] - d["MACD_Signal"]

    d["MA_10"]       = d["Close"].rolling(10).mean()
    d["MA_50"]       = d["Close"].rolling(50).mean()
    # +ve = short MA above long MA (bullish golden cross)
    d["MA_Cross"]    = d["MA_10"] - d["MA_50"]

    # ── MOMENTUM ────────────────────────────────────────────────
    delta         = d["Close"].diff()
    gain          = delta.clip(lower=0)
    loss          = -delta.clip(upper=0)
    avg_gain      = gain.rolling(14).mean()
    avg_loss      = loss.rolling(14).mean()
    rs            = avg_gain / (avg_loss + 1e-10)
    d["RSI"]      = 100 - (100 / (1 + rs))
    d["RSI_Norm"] = d["RSI"] / 100.0              # bounded [0, 1] ✅

    d["ROC_5"]    = d["Close"].pct_change(5)  * 100
    d["ROC_10"]   = d["Close"].pct_change(10) * 100
    d["ROC_20"]   = d["Close"].pct_change(20) * 100

    # Stochastic %K and %D
    low_14        = d["Low"].rolling(14).min()
    high_14       = d["High"].rolling(14).max()
    d["Stoch_K"]  = ((d["Close"] - low_14) /
                     (high_14 - low_14 + 1e-10) * 100)
    d["Stoch_D"]  = d["Stoch_K"].rolling(3).mean()

    # ── VOLATILITY ──────────────────────────────────────────────
    d["BB_Mid"]   = d["Close"].rolling(20).mean()
    d["BB_Std"]   = d["Close"].rolling(20).std()
    d["BB_Upper"] = d["BB_Mid"] + 2 * d["BB_Std"]
    d["BB_Lower"] = d["BB_Mid"] - 2 * d["BB_Std"]
    # normalised width → stationary ✅
    d["BB_Width"] = ((d["BB_Upper"] - d["BB_Lower"]) /
                     (d["BB_Mid"] + 1e-10))
    # %B: 0=at lower band, 1=at upper band ✅
    d["BB_Pct"]   = ((d["Close"]   - d["BB_Lower"]) /
                     (d["BB_Upper"] - d["BB_Lower"] + 1e-10))

    hl           = d["High"] - d["Low"]
    hc           = (d["High"] - d["Close"].shift(1)).abs()
    lc           = (d["Low"]  - d["Close"].shift(1)).abs()
    tr           = pd.concat([hl, hc, lc], axis=1).max(axis=1)
    d["ATR"]     = tr.rolling(14).mean()
    d["ATR_Pct"] = d["ATR"] / (d["Close"] + 1e-10)   # normalised ✅

    # Annualised historical volatility
    d["HV_20"]   = (d["Close"].pct_change()
                               .rolling(20).std() * np.sqrt(252))

    # ── VOLUME ──────────────────────────────────────────────────
    # On-Balance Volume
    obv = [0]
    for i in range(1, len(d)):
        if   d["Close"].iloc[i] > d["Close"].iloc[i - 1]:
            obv.append(obv[-1] + d["Volume"].iloc[i])
        elif d["Close"].iloc[i] < d["Close"].iloc[i - 1]:
            obv.append(obv[-1] - d["Volume"].iloc[i])
        else:
            obv.append(obv[-1])
    d["OBV"]        = obv
    d["OBV_EMA"]    = (pd.Series(obv, index=d.index)
                         .ewm(span=10).mean().values)
    d["Vol_MA20"]   = d["Volume"].rolling(20).mean()
    d["Vol_Ratio"]  = d["Volume"] / (d["Vol_MA20"] + 1e-10)  # ✅

    # ── PRICE STRUCTURE ─────────────────────────────────────────
    d["HL_Pct"]       = (d["High"] - d["Low"]) / (d["Close"] + 1e-10)
    d["Gap_Pct"]      = ((d["Open"] - d["Close"].shift(1)) /
                          (d["Close"].shift(1) + 1e-10))
    d["Upper_Shadow"] = ((d["High"] - d[["Open","Close"]].max(axis=1)) /
                          (d["Close"] + 1e-10))
    d["Lower_Shadow"] = ((d[["Open","Close"]].min(axis=1) - d["Low"]) /
                          (d["Close"] + 1e-10))

    d.dropna(inplace=True)

    print(f"  Raw columns     : 5  (OHLCV)")
    print(f"  New indicators  : {len(d.columns) - 5}")
    print(f"  Total columns   : {len(d.columns)}")
    print(f"  Rows after drop : {len(d):,}")
    return d


# ================================================================
# STEP 3 — STATIONARY FEATURES   (Concept 2 — pct_change)
# ================================================================
#
#   WHY: Raw price is non-stationary — it drifts upward over years.
#        A model trained on ₹50 prices won't generalize to ₹1400.
#        Returns (% changes) have a stable mean ~0 and consistent
#        variance → LSTM learns patterns, not absolute price levels.
#
#   ORDER:
#     1. Compute all indicators from raw prices  (Step 2 already done)
#     2. Convert raw OHLCV → returns            (this step)
#     3. Drop raw price columns
#
def make_stationary(df: pd.DataFrame):
    """
    Returns (df_stat, features_list).
    Last column of df_stat is always Target_Return.
    """
    print("\n" + "="*65)
    print("  STEP 3 — Stationarity via pct_change  (Concept 2)")
    print("="*65)

    d = df.copy()

    # Raw price columns → percentage returns
    d["Open_Return"]  = d["Open"].pct_change()
    d["High_Return"]  = d["High"].pct_change()
    d["Low_Return"]   = d["Low"].pct_change()
    d["Close_Return"] = d["Close"].pct_change()

    # Volume → log-change (handles large volume spikes better)
    # Volume already has no zeros (cleaned in Step 1)
    d["Volume_Log"]   = np.log(d["Volume"] / d["Volume"].shift(1))

    # ── TARGET: next day's Close return ─────────────────────────
    d["Target_Return"] = d["Close"].pct_change().shift(-1)

    # ── Drop raw non-stationary columns ─────────────────────────
    # (indicators have already been extracted from them in Step 2)
    drop_cols = [
        "Open", "High", "Low", "Close", "Volume",   # raw OHLCV
        "EMA_12", "EMA_26",                          # intermediate
        "MA_10",  "MA_50",                           # intermediate
        "BB_Mid", "BB_Std", "BB_Upper", "BB_Lower",  # intermediate
        "ATR", "OBV", "Vol_MA20",                    # intermediate / non-stationary
    ]
    d.drop(columns=[c for c in drop_cols if c in d.columns],
           inplace=True)
    d.dropna(inplace=True)

    # ── Final ordered feature list ───────────────────────────────
    features = [
        # Converted price (stationary returns)
        "Open_Return",  "High_Return",   "Low_Return",
        "Close_Return", "Volume_Log",

        # Trend (oscillate near 0 — already stationary)
        "MACD", "MACD_Signal", "MACD_Hist", "MA_Cross",

        # Momentum (bounded / stationary)
        "RSI_Norm", "ROC_5", "ROC_10", "ROC_20",
        "Stoch_K",  "Stoch_D",

        # Volatility (ratio / normalised — stationary)
        "BB_Width", "BB_Pct", "ATR_Pct", "HV_20",

        # Volume (ratio — stationary)
        "Vol_Ratio", "OBV_EMA",

        # Price structure (ratio — stationary)
        "HL_Pct", "Gap_Pct", "Upper_Shadow", "Lower_Shadow",
    ]
    features = [f for f in features if f in d.columns]

    print(f"\n  Features ({len(features)}):")
    for i, f in enumerate(features, 1):
        print(f"    {i:>2}. {f}")
    print(f"\n  Target  : {TARGET}  (next-day Close return)")
    print(f"  Rows    : {len(d):,}")

    return d, features


# ================================================================
# STEP 4 — SEQUENCE CREATION   (Multi-Step y)
# ================================================================
def create_sequences(data: np.ndarray,
                     lookback: int,
                     horizon: int):
    """
    Sliding window over a 2-D scaled array.

    Convention:
        All columns except the last = features
        Last column                 = target

    Returns
    -------
    X : ndarray  (samples, lookback, n_features)
    y : ndarray  (samples, horizon)              ← multi-step
    """
    X, y = [], []
    for i in range(lookback, len(data) - horizon + 1):
        X.append(data[i - lookback : i,  :-1])     # feature window
        y.append(data[i : i + horizon,    -1])      # next N targets
    return np.array(X), np.array(y)


# ================================================================
# STEP 5 — BIDIRECTIONAL LSTM MODEL
# ================================================================
def build_model(lookback: int,
                n_features: int,
                horizon: int) -> tf.keras.Model:
    """
    Architecture
    ─────────────────────────────────────────────────
    BiLSTM(128) return_sequences=True
        Reads sequence forward AND backward.
        return_sequences=True passes full seq to next layer.
    Dropout(0.2)

    BiLSTM(64)  return_sequences=True
        Second bidirectional pass — deeper patterns.
    Dropout(0.2)

    LSTM(32)    return_sequences=False
        Collapses full sequence → single context vector.
    Dropout(0.2)

    Dense(64, relu) → Dense(32, relu) → Dense(horizon)
        Outputs all N future days simultaneously.
    ─────────────────────────────────────────────────
    Loss: Huber — less sensitive to return outliers than MSE.
    """
    model = Sequential([
        # Layer 1 — BiLSTM, return full sequence
        Bidirectional(
            LSTM(128, return_sequences=True,
                 dropout=0.1, recurrent_dropout=0.1),
            input_shape=(lookback, n_features)
        ),
        Dropout(0.2),

        # Layer 2 — BiLSTM, still full sequence
        Bidirectional(
            LSTM(64, return_sequences=True,
                 dropout=0.1, recurrent_dropout=0.1)
        ),
        Dropout(0.2),

        # Layer 3 — collapse sequence → context vector
        LSTM(32, return_sequences=False),
        Dropout(0.2),

        # Dense head — all N horizon days at once
        Dense(64, activation="relu"),
        Dropout(0.1),
        Dense(32, activation="relu"),
        Dense(horizon),                 # N output neurons
    ])

    model.compile(
        optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
        loss="huber",
        metrics=["mae"]
    )
    return model


def _get_callbacks():
    return [
        EarlyStopping(monitor="val_loss", patience=15,
                      restore_best_weights=True, verbose=1),
        ReduceLROnPlateau(monitor="val_loss", factor=0.5,
                          patience=7, min_lr=1e-6, verbose=0),
    ]


# ================================================================
# STEP 6 — WALK-FORWARD VALIDATION   (Concept 3)
# ================================================================
#
#   WHY: A single 80/20 split evaluates the model on ONE window.
#        If that window happens to be a bull run, metrics look great.
#        If it's a crash, metrics look terrible.
#        Walk-forward tests across N non-overlapping windows →
#        a much more honest picture of real-world performance.
#
#   Expanding window — training set grows each fold:
#
#   Fold 1: [=Train=]            |Test|
#   Fold 2: [=====Train=====]         |Test|
#   Fold 3: [=========Train=========]      |Test|
#   ...
#
#   Key rule: scaler is re-fitted on each fold's TRAIN set only.
#
def walk_forward_validation(df_stat: pd.DataFrame,
                             features: list):
    print("\n" + "="*65)
    print("  STEP 6 — Walk-Forward Validation  (Concept 3)")
    print("="*65)

    all_cols  = features + [TARGET]
    df_clean  = df_stat[all_cols].dropna()
    data      = df_clean.values
    total     = len(data)
    test_size = total // (N_SPLITS + 1)

    print(f"  Total rows    : {total:,}")
    print(f"  Test window   : {test_size:,} rows / fold")
    print(f"  Folds         : {N_SPLITS}")

    fold_records  = []
    best_val_loss = np.inf
    best_model    = None
    best_scaler   = None
    best_tgt_scl  = None

    for fold in range(N_SPLITS):

        train_end  = test_size * (fold + 1)
        test_start = train_end
        test_end   = test_start + test_size
        if test_end > total:
            break

        tr_data  = data[:train_end]
        te_data  = data[test_start : test_end]
        tr_dates = df_clean.index[:train_end]
        te_dates = df_clean.index[test_start : test_end]

        print(f"\n  ── Fold {fold + 1}/{N_SPLITS} ──")
        print(f"     Train : {tr_dates[0].date()} → {tr_dates[-1].date()}"
              f"  ({len(tr_data):,} rows)")
        print(f"     Test  : {te_dates[0].date()} → {te_dates[-1].date()}"
              f"  ({len(te_data):,} rows)")

        # ── Scale on TRAIN only (prevent leakage) ─────────────
        scaler     = MinMaxScaler()
        tr_scaled  = scaler.fit_transform(tr_data)   # fit + transform
        te_scaled  = scaler.transform(te_data)        # transform only

        tgt_scaler = MinMaxScaler()
        tgt_scaler.fit(tr_data[:, -1].reshape(-1, 1))

        # ── Sequences ──────────────────────────────────────────
        X_tr, y_tr = create_sequences(tr_scaled, LOOKBACK, FORECAST_HORIZON)
        X_te, y_te = create_sequences(te_scaled, LOOKBACK, FORECAST_HORIZON)

        if len(X_tr) == 0 or len(X_te) == 0:
            print("     Skipping — not enough rows")
            continue

        # ── Fresh model each fold ──────────────────────────────
        set_seed(42)
        model = build_model(LOOKBACK, X_tr.shape[2], FORECAST_HORIZON)

        history = model.fit(
            X_tr, y_tr,
            epochs=EPOCHS,
            batch_size=BATCH_SIZE,
            validation_split=0.1,
            shuffle=False,               # MUST be False — time order
            callbacks=_get_callbacks(),
            verbose=0
        )

        stopped   = len(history.history["loss"])
        val_loss  = min(history.history["val_loss"])

        # ── Evaluate ───────────────────────────────────────────
        yp_sc   = model.predict(X_te, verbose=0)
        y_pred  = tgt_scaler.inverse_transform(
                      yp_sc.reshape(-1, 1)
                  ).reshape(-1, FORECAST_HORIZON)
        y_act   = tgt_scaler.inverse_transform(
                      y_te.reshape(-1, 1)
                  ).reshape(-1, FORECAST_HORIZON)

        mae  = mean_absolute_error(y_act.flatten(), y_pred.flatten())
        rmse = np.sqrt(mean_squared_error(y_act.flatten(), y_pred.flatten()))
        mape = np.mean(
                   np.abs((y_act.flatten() - y_pred.flatten()) /
                           (np.abs(y_act.flatten()) + 1e-10))
               ) * 100
        dacc = np.mean(
                   np.sign(y_act.flatten()) ==
                   np.sign(y_pred.flatten())
               ) * 100

        print(f"     Epochs={stopped}  val_loss={val_loss:.6f}")
        print(f"     MAE={mae:.6f}  RMSE={rmse:.6f}"
              f"  MAPE={mape:.2f}%  DirAcc={dacc:.1f}%")

        fold_records.append({
            "fold":       fold + 1,
            "train_rows": len(tr_data),
            "train_from": str(tr_dates[0].date()),
            "train_to":   str(tr_dates[-1].date()),
            "test_from":  str(te_dates[0].date()),
            "test_to":    str(te_dates[-1].date()),
            "epochs":     stopped,
            "val_loss":   round(val_loss, 6),
            "MAE":        round(mae,  6),
            "RMSE":       round(rmse, 6),
            "MAPE":       round(mape, 2),
            "DirAcc":     round(dacc, 1),
        })

        # Track best model across all folds
        if val_loss < best_val_loss:
            best_val_loss = val_loss
            best_model    = model
            best_scaler   = scaler
            best_tgt_scl  = tgt_scaler
            print(f"     ★  New best model  (val_loss={val_loss:.6f})")

    # ── Summary table ─────────────────────────────────────────
    results = pd.DataFrame(fold_records)
    print("\n" + "="*65)
    print("  Walk-Forward Summary")
    print("="*65)
    print(results[["fold","MAE","RMSE","MAPE","DirAcc","epochs"]]
          .to_string(index=False))
    print("-"*65)
    print(f"  Avg MAE    : {results['MAE'].mean():.6f}"
          f"  ±{results['MAE'].std():.6f}")
    print(f"  Avg RMSE   : {results['RMSE'].mean():.6f}"
          f"  ±{results['RMSE'].std():.6f}")
    print(f"  Avg MAPE   : {results['MAPE'].mean():.2f}%"
          f"  ±{results['MAPE'].std():.2f}%")
    print(f"  Avg DirAcc : {results['DirAcc'].mean():.1f}%"
          f"  ±{results['DirAcc'].std():.1f}%")

    return results, best_model, best_scaler, best_tgt_scl


# ================================================================
# STEP 7 — FINAL FORECAST ON UNSEEN FUTURE DATA
# ================================================================
def final_forecast(df_stat: pd.DataFrame,
                   df_raw:  pd.DataFrame,
                   features: list,
                   best_model,
                   best_scaler,
                   best_tgt_scl):
    """
    Uses the best model from walk-forward validation to predict
    the next FORECAST_HORIZON business days.

    Process:
      1. Scale full dataset with best_scaler (transform only — no refit)
      2. Take last LOOKBACK rows as the live input window
      3. Predict returns for all N days in one forward pass
      4. Compound predicted returns → INR price series
    """
    print("\n" + "="*65)
    print(f"  STEP 7 — Final Forecast  ({FORECAST_HORIZON} days ahead)")
    print("="*65)

    all_cols    = features + [TARGET]
    df_clean    = df_stat[all_cols].dropna()
    full_scaled = best_scaler.transform(df_clean.values)  # transform only

    # Last LOOKBACK rows, features only (drop target column)
    last_seq    = full_scaled[-LOOKBACK:, :-1]
    inp         = last_seq.reshape(1, LOOKBACK, len(features))

    pred_sc     = best_model.predict(inp, verbose=0)      # (1, N)
    pred_ret    = best_tgt_scl.inverse_transform(
                      pred_sc.reshape(-1, 1)
                  ).flatten()

    # Compound returns → INR prices
    last_price  = float(df_raw["Close"].iloc[-1])
    prices      = [last_price]
    for r in pred_ret:
        prices.append(prices[-1] * (1 + r))
    pred_prices = np.array(prices[1:])

    future_dates = pd.bdate_range(
        start=df_raw.index[-1] + pd.Timedelta(days=1),
        periods=FORECAST_HORIZON
    )

    result = pd.DataFrame({
        "Predicted_Return" : pred_ret.round(6),
        "Predicted_Close"  : pred_prices.round(2),
        "Change_Pct"       : (pred_ret * 100).round(3),
    }, index=future_dates)
    result.index.name = "Date"

    print(f"\n  Last known Close : ₹{last_price:.2f}")
    print(f"\n  {'Date':<14}  {'Return':>10}  {'Close (₹)':>12}  Dir")
    print(f"  {'─'*50}")
    for dt, row in result.iterrows():
        arrow = "▲" if row["Predicted_Return"] >= 0 else "▼"
        print(f"  {str(dt.date()):<14}"
              f"  {row['Predicted_Return']:>+9.4%}"
              f"  ₹{row['Predicted_Close']:>10.2f}"
              f"  {arrow}")

    return result, pred_ret, pred_prices, future_dates


# ================================================================
# STEP 8 — PLOT EVERYTHING
# ================================================================
def plot_everything(df_raw:      pd.DataFrame,
                    wf_results:  pd.DataFrame,
                    result_df:   pd.DataFrame,
                    pred_prices: np.ndarray,
                    pred_ret:    np.ndarray,
                    future_dates,
                    features:    list,
                    best_model,
                    best_scaler,
                    best_tgt_scl,
                    df_stat:     pd.DataFrame):
    """
    Dark-theme 7-panel dashboard:
      P1 (top, wide) — Historical close + forecast ribbon
      P2             — Walk-forward RMSE per fold
      P3             — Walk-forward Direction Accuracy per fold
      P4             — Walk-forward MAPE per fold
      P5             — Predicted future daily returns
      P6             — Per-day error & direction accuracy (best fold)
      P7             — Recent 200-day volume with MA20
    """
    # ── Colour palette ─────────────────────────────────────────
    DARK   = "#0D1117"
    CARD   = "#161B22"
    TEXT   = "#C9D1D9"
    MUTED  = "#6E7681"
    BLUE   = "#58A6FF"
    ORANGE = "#F78166"
    GREEN  = "#3FB950"
    RED    = "#F85149"
    GOLD   = "#E3B341"
    PURPLE = "#BC8CFF"

    fig = plt.figure(figsize=(22, 18))
    fig.patch.set_facecolor(DARK)
    gs  = gridspec.GridSpec(3, 3, figure=fig,
                            hspace=0.52, wspace=0.35)

    def style(ax, title, ylabel=""):
        ax.set_facecolor(CARD)
        ax.set_title(title, color=TEXT, fontsize=10,
                     fontweight="bold", pad=8)
        ax.tick_params(colors=TEXT, labelsize=8)
        ax.xaxis.label.set_color(TEXT)
        ax.yaxis.label.set_color(TEXT)
        if ylabel:
            ax.set_ylabel(ylabel, color=TEXT, fontsize=9)
        for sp in ax.spines.values():
            sp.set_edgecolor("#30363D")
        ax.grid(linestyle="--", alpha=0.18, color=MUTED)

    # ── P1: History + forecast (top, spans all 3 cols) ─────────
    ax1 = fig.add_subplot(gs[0, :])
    n   = min(600, len(df_raw))
    rec = df_raw["Close"].iloc[-n:]

    ax1.plot(rec.index, rec.values,
             color=BLUE, lw=1.4,
             label="Historical Close", alpha=0.9)

    unc = np.array([abs(pred_prices[i]) * (0.007 + i * 0.0018)
                    for i in range(len(pred_prices))])
    ax1.plot(future_dates, pred_prices,
             color=ORANGE, lw=2.8, ls="--",
             label=f"{FORECAST_HORIZON}-Day Forecast", zorder=5)
    ax1.fill_between(future_dates,
                     pred_prices - unc,
                     pred_prices + unc,
                     color=ORANGE, alpha=0.20,
                     label="Uncertainty band")
    ax1.axvline(df_raw.index[-1], color=GOLD,
                ls=":", lw=1.5, alpha=0.85)
    ax1.axvspan(df_raw.index[-1], future_dates[-1],
                alpha=0.06, color=ORANGE)

    last_c = float(df_raw["Close"].iloc[-1])
    ax1.annotate(f"₹{last_c:.0f}",
                 xy=(df_raw.index[-1], last_c),
                 xytext=(-60, 20), textcoords="offset points",
                 color=BLUE, fontsize=9,
                 arrowprops=dict(arrowstyle="->",
                                 color=BLUE, lw=1.2))
    ax1.annotate(f"₹{pred_prices[-1]:.0f}",
                 xy=(future_dates[-1], pred_prices[-1]),
                 xytext=(10, 20), textcoords="offset points",
                 color=ORANGE, fontsize=9,
                 arrowprops=dict(arrowstyle="->",
                                 color=ORANGE, lw=1.2))

    style(ax1, "RELIANCE — Historical Close + Future Forecast",
          "Price (₹ INR)")
    ax1.legend(fontsize=8, facecolor=CARD,
               labelcolor=TEXT, loc="upper left")
    ax1.xaxis.set_major_formatter(mdates.DateFormatter("%b %Y"))
    ax1.xaxis.set_major_locator(mdates.MonthLocator(interval=3))
    plt.setp(ax1.xaxis.get_majorticklabels(),
             rotation=28, ha="right", color=TEXT, fontsize=8)

    folds = wf_results["fold"].astype(str)

    # ── P2: RMSE per fold ───────────────────────────────────────
    ax2   = fig.add_subplot(gs[1, 0])
    rmses = wf_results["RMSE"].values
    avg_r = rmses.mean()
    c2    = [GREEN if v <= avg_r else ORANGE for v in rmses]
    bars2 = ax2.bar(folds, rmses, color=c2, width=0.5, alpha=0.85)
    ax2.axhline(avg_r, color=GOLD, ls="--", lw=1.4,
                label=f"Avg {avg_r:.4f}")
    for b, v in zip(bars2, rmses):
        ax2.text(b.get_x() + b.get_width() / 2,
                 v + avg_r * 0.02, f"{v:.4f}",
                 ha="center", va="bottom",
                 fontsize=7, color=TEXT)
    style(ax2, "Walk-Forward RMSE / Fold", "RMSE")
    ax2.set_xlabel("Fold", color=TEXT, fontsize=9)
    ax2.legend(fontsize=8, facecolor=CARD, labelcolor=TEXT)

    # ── P3: Direction Accuracy per fold ─────────────────────────
    ax3   = fig.add_subplot(gs[1, 1])
    daccs = wf_results["DirAcc"].values
    avg_d = daccs.mean()
    c3    = [GREEN if v >= avg_d else ORANGE for v in daccs]
    bars3 = ax3.bar(folds, daccs, color=c3, width=0.5, alpha=0.85)
    ax3.axhline(avg_d, color=GOLD, ls="--", lw=1.4,
                label=f"Avg {avg_d:.1f}%")
    ax3.axhline(50, color=RED, ls=":", lw=1.0,
                label="50% baseline", alpha=0.7)
    for b, v in zip(bars3, daccs):
        ax3.text(b.get_x() + b.get_width() / 2,
                 v + 0.5, f"{v:.1f}%",
                 ha="center", va="bottom",
                 fontsize=7, color=TEXT)
    style(ax3, "Walk-Forward Direction Accuracy / Fold",
          "Accuracy (%)")
    ax3.set_xlabel("Fold", color=TEXT, fontsize=9)
    ax3.legend(fontsize=8, facecolor=CARD, labelcolor=TEXT)

    # ── P4: MAPE per fold ───────────────────────────────────────
    ax4   = fig.add_subplot(gs[1, 2])
    mapes = wf_results["MAPE"].values
    avg_m = mapes.mean()
    c4    = [GREEN if v <= avg_m else ORANGE for v in mapes]
    bars4 = ax4.bar(folds, mapes, color=c4, width=0.5, alpha=0.85)
    ax4.axhline(avg_m, color=GOLD, ls="--", lw=1.4,
                label=f"Avg {avg_m:.2f}%")
    for b, v in zip(bars4, mapes):
        ax4.text(b.get_x() + b.get_width() / 2,
                 v + avg_m * 0.02, f"{v:.2f}%",
                 ha="center", va="bottom",
                 fontsize=7, color=TEXT)
    style(ax4, "Walk-Forward MAPE / Fold", "MAPE (%)")
    ax4.set_xlabel("Fold", color=TEXT, fontsize=9)
    ax4.legend(fontsize=8, facecolor=CARD, labelcolor=TEXT)

    # ── P5: Predicted future returns (bar) ──────────────────────
    ax5  = fig.add_subplot(gs[2, 0])
    rpct = result_df["Change_Pct"].values
    c5   = [GREEN if v >= 0 else RED for v in rpct]
    bars5 = ax5.bar(range(1, len(rpct) + 1),
                    rpct, color=c5, width=0.6, alpha=0.85)
    ax5.axhline(0, color=MUTED, lw=0.8)
    for b, v in zip(bars5, rpct):
        ax5.text(b.get_x() + b.get_width() / 2,
                 v + np.sign(v) * 0.012,
                 f"{v:+.2f}%", ha="center",
                 va="bottom" if v >= 0 else "top",
                 fontsize=7, color=TEXT)
    style(ax5,
          f"Predicted Daily Returns — Next {FORECAST_HORIZON} Days",
          "Return (%)")
    ax5.set_xlabel("Forecast Day", color=TEXT, fontsize=9)
    ax5.set_xticks(range(1, len(rpct) + 1))

    # ── P6: Per-day error degradation (best fold data) ──────────
    ax6 = fig.add_subplot(gs[2, 1])
    all_cols  = features + [TARGET]
    df_clean  = df_stat[all_cols].dropna()
    total     = len(df_clean)
    test_size = total // (N_SPLITS + 1)
    te_raw    = df_clean.values[-test_size:]
    te_sc     = best_scaler.transform(te_raw)
    X_pd, y_pd = create_sequences(te_sc, LOOKBACK, FORECAST_HORIZON)

    if len(X_pd) > 0:
        yp_sc2  = best_model.predict(X_pd, verbose=0)
        yp2     = best_tgt_scl.inverse_transform(
                      yp_sc2.reshape(-1, 1)
                  ).reshape(-1, FORECAST_HORIZON)
        ya2     = best_tgt_scl.inverse_transform(
                      y_pd.reshape(-1, 1)
                  ).reshape(-1, FORECAST_HORIZON)

        days_  = list(range(1, FORECAST_HORIZON + 1))
        maes_  = [mean_absolute_error(ya2[:, d], yp2[:, d])
                  for d in range(FORECAST_HORIZON)]
        rmses_ = [np.sqrt(mean_squared_error(ya2[:, d], yp2[:, d]))
                  for d in range(FORECAST_HORIZON)]
        dirs_  = [np.mean(np.sign(ya2[:, d]) ==
                           np.sign(yp2[:, d])) * 100
                  for d in range(FORECAST_HORIZON)]

        ax6b = ax6.twinx()
        ax6.plot(days_, maes_,  color=BLUE,   lw=2,
                 marker="o", ms=5, label="MAE")
        ax6.plot(days_, rmses_, color=ORANGE,  lw=2,
                 marker="s", ms=5, ls="--", label="RMSE")
        ax6b.plot(days_, dirs_, color=GREEN,   lw=2,
                  marker="^", ms=5, ls=":", label="DirAcc%")
        ax6b.set_ylabel("Direction Accuracy (%)",
                         color=GREEN, fontsize=9)
        ax6b.tick_params(colors=GREEN, labelsize=8)
        l1, lb1 = ax6.get_legend_handles_labels()
        l2, lb2 = ax6b.get_legend_handles_labels()
        ax6.legend(l1 + l2, lb1 + lb2,
                   fontsize=7, facecolor=CARD, labelcolor=TEXT)

    style(ax6, "Per-Day Error (Best Fold)", "MAE / RMSE")
    ax6.set_xlabel("Forecast Day", color=TEXT, fontsize=9)

    # ── P7: Recent volume ────────────────────────────────────────
    ax7  = fig.add_subplot(gs[2, 2])
    vrec = df_raw["Volume"].iloc[-200:]
    ax7.bar(vrec.index, vrec.values,
            color=PURPLE, width=1.5, alpha=0.65)
    vma  = vrec.rolling(20).mean()
    ax7.plot(vma.index, vma.values,
             color=GOLD, lw=1.5, label="20d MA")
    style(ax7, "Recent Volume (200 days)", "Volume")
    ax7.set_xlabel("Date", color=TEXT, fontsize=9)
    ax7.yaxis.set_major_formatter(
        plt.FuncFormatter(lambda x, _: f"{x/1e6:.0f}M"))
    ax7.xaxis.set_major_formatter(mdates.DateFormatter("%b %y"))
    ax7.xaxis.set_major_locator(mdates.MonthLocator(interval=2))
    plt.setp(ax7.xaxis.get_majorticklabels(),
             rotation=28, ha="right",
             color=TEXT, fontsize=7)
    ax7.legend(fontsize=8, facecolor=CARD, labelcolor=TEXT)

    # ── Super-title ──────────────────────────────────────────────
    fig.suptitle(
        f"RELIANCE  |  BiLSTM Multi-Step Forecast  |  "
        f"Lookback={LOOKBACK}d  Horizon={FORECAST_HORIZON}d  "
        f"Features={len(features)}  Folds={N_SPLITS}",
        fontsize=13, fontweight="bold",
        color=TEXT, y=1.01
    )

    out = "RELIANCE_bilstm_dashboard.png"
    plt.savefig(out, dpi=150, bbox_inches="tight",
                facecolor=fig.get_facecolor())
    plt.show()
    print(f"\n  Dashboard saved  →  {out}")
    return out


# ================================================================
# STEP 9 — SAVE EVERYTHING
# ================================================================
def save_everything(best_model,
                    best_scaler,
                    best_tgt_scl,
                    features:   list,
                    wf_results: pd.DataFrame,
                    result_df:  pd.DataFrame):
    """
    Saves 6 artefacts:
      *.keras         trained model weights
      *_scaler.pkl    full-feature MinMaxScaler (fitted on best-fold train)
      *_target_scaler target-only MinMaxScaler
      *_features.csv  ordered feature list (25 names)
      *_wf_results    walk-forward fold metrics
      *_forecast.csv  final N-day forecast table

    To reload later:
      from tensorflow.keras.models import load_model
      import joblib
      model  = load_model("RELIANCE_bilstm_model.keras")
      scaler = joblib.load("RELIANCE_scaler.pkl")
    """
    print("\n" + "="*65)
    print("  STEP 9 — Save Everything")
    print("="*65)

    best_model.save("RELIANCE_bilstm_model.keras")
    joblib.dump(best_scaler,  "RELIANCE_scaler.pkl")
    joblib.dump(best_tgt_scl, "RELIANCE_target_scaler.pkl")
    pd.Series(features).to_csv(
        "RELIANCE_features.csv", index=False, header=["feature"])
    wf_results.to_csv("RELIANCE_wf_results.csv", index=False)
    result_df.to_csv("RELIANCE_forecast.csv")

    saved = [
        "RELIANCE_bilstm_model.keras",
        "RELIANCE_scaler.pkl",
        "RELIANCE_target_scaler.pkl",
        "RELIANCE_features.csv",
        "RELIANCE_wf_results.csv",
        "RELIANCE_forecast.csv",
        "RELIANCE_bilstm_dashboard.png",
    ]
    for f in saved:
        print(f"  ✓  {f}")


# ================================================================
#   MAIN — Wire all 9 steps together
# ================================================================
if __name__ == "__main__":

    # ── Step 1 ───────────────────────────────────────────────────
    df_raw = load_data(CSV_PATH)

    # ── Step 2 ───────────────────────────────────────────────────
    df_ind = add_technical_indicators(df_raw)

    # ── Step 3 ───────────────────────────────────────────────────
    df_stat, features = make_stationary(df_ind)

    # ── Step 4 — sequence shape preview ─────────────────────────
    print("\n" + "="*65)
    print("  STEP 4 — Sequence Shape Preview")
    print("="*65)
    _dc = df_stat[features + [TARGET]].dropna()
    _sc = MinMaxScaler()
    _tmp = _sc.fit_transform(_dc.values)
    _X, _y = create_sequences(_tmp, LOOKBACK, FORECAST_HORIZON)
    print(f"  X shape : {_X.shape}   (samples, lookback, n_features)")
    print(f"  y shape : {_y.shape}   (samples, horizon)  ← multi-step")
    del _sc, _tmp, _X, _y, _dc

    # ── Step 5 — model preview ───────────────────────────────────
    print("\n" + "="*65)
    print("  STEP 5 — Model Architecture")
    print("="*65)
    _m = build_model(LOOKBACK, len(features), FORECAST_HORIZON)
    _m.summary()
    del _m

    # ── Step 6 — walk-forward validation ────────────────────────
    wf_results, best_model, best_scaler, best_tgt_scl = \
        walk_forward_validation(df_stat, features)

    # ── Step 7 — forecast future ────────────────────────────────
    result_df, pred_ret, pred_prices, future_dates = \
        final_forecast(df_stat, df_raw, features,
                       best_model, best_scaler, best_tgt_scl)

    # ── Step 8 — plot everything ─────────────────────────────────
    plot_everything(
        df_raw, wf_results, result_df,
        pred_prices, pred_ret, future_dates,
        features, best_model, best_scaler,
        best_tgt_scl, df_stat
    )

    # ── Step 9 — save everything ─────────────────────────────────
    save_everything(
        best_model, best_scaler, best_tgt_scl,
        features, wf_results, result_df
    )

    print("\n" + "="*65)
    print("  Pipeline Complete!")
    print("="*65)
```
