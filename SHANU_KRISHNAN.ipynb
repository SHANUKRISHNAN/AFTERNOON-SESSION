```python
x=df.drop(['Loan_Status','Loan_ID'],axis=1)
y=df['Loan_Status']
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=41)
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()

x_train['Gender']=le.fit_transform(x_train['Gender'])
x_test['Gender']=le.transform(x_test['Gender'])

x_train['Married']=le.fit_transform(x_train['Married'])
x_test['Married']=le.transform(x_test['Married'])

x_train['Education']=le.fit_transform(x_train['Education'])
x_test['Education']=le.transform(x_test['Education'])

x_train['Self_Employed']=le.fit_transform(x_train['Self_Employed'])
x_test['Self_Employed']=le.transform(x_test['Self_Employed'])

y_train=le.fit_transform(y_train)
y_test=le.transform(y_test)
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe=OneHotEncoder(handle_unknown='ignore', sparse_output=False)

x_train_prop_area_encoded = ohe.fit_transform(x_train[['Property_Area']])
x_test_prop_area_encoded = ohe.transform(x_test[['Property_Area']])

feature_names = ohe.get_feature_names_out(['Property_Area'])

x_train_prop_area_df = pd.DataFrame(x_train_prop_area_encoded, columns=feature_names, index=x_train.index)
x_test_prop_area_df = pd.DataFrame(x_test_prop_area_encoded, columns=feature_names, index=x_test.index)

x_train = pd.concat([x_train.drop('Property_Area', axis=1), x_train_prop_area_df], axis=1)
x_test = pd.concat([x_test.drop('Property_Area', axis=1), x_test_prop_area_df], axis=1)
```

```python
x_train['Dependents']=x_train['Dependents'].replace('3+',3)
x_train['Dependents']=x_train['Dependents'].replace('mode4',4)
x_train['Dependents'] = x_train['Dependents'].astype(int)

x_test['Dependents']=x_test['Dependents'].replace('3+',3)
x_test['Dependents']=x_test['Dependents'].replace('mode4',4)
x_test['Dependents'] = x_test['Dependents'].astype(int)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Input
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()

```

```python
x_train_numerical = x_train[['ApplicantIncome','CoapplicantIncome','LoanAmount','Loan_Amount_Term']]
x_test_numerical = x_test[['ApplicantIncome','CoapplicantIncome','LoanAmount','Loan_Amount_Term']]

x_train_scaled = ss.fit_transform(x_train_numerical)
x_test_scaled = ss.transform(x_test_numerical)

x_train[['ApplicantIncome','CoapplicantIncome','LoanAmount','Loan_Amount_Term']] = x_train_scaled
x_test[['ApplicantIncome','CoapplicantIncome','LoanAmount','Loan_Amount_Term']] = x_test_scaled
```

```python
model=Sequential()
model.add(Input(shape=(13,)))
model.add(Dense(32,activation='relu'))
model.add(Dense(16,activation='relu'))
model.add(Dense(8,activation='relu'))

model.add(Dense(2,activation='sigmoid')) # binary
```

```python
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])

learned=model.fit(x_train,y_train,epochs=100,batch_size=10,verbose=1,validation_data=(x_test,y_test))

loss, accuracy=model.evaluate(x_test,y_test)
print('Test Loss:',loss)
print('Test Accuracy:',accuracy)

```
