```python
def add_features(df: pd.DataFrame) -> pd.DataFrame:
    # ── Stationary returns ───────────────────────────────────────────────
    df["Return"]      = df["Close"].pct_change()
    df["Log_Return"]  = np.log(df["Close"] / df["Close"].shift(1))

    # ── Intraday structure ───────────────────────────────────────────────
    df["HL_Ratio"]    = (df["High"] - df["Low"]) / df["Close"]
    df["OC_Ratio"]    = (df["Close"] - df["Open"]) / df["Open"]

    # ── Momentum ─────────────────────────────────────────────────────────
    df["Momentum_5"]  = df["Close"] - df["Close"].shift(5)
    df["Momentum_10"] = df["Close"] - df["Close"].shift(10)

    # ── Realized volatility ──────────────────────────────────────────────
    df["Vol_10"]      = df["Return"].rolling(10).std()

    # ── Trend signal: ratio only (not raw MA values → no leakage) ────────
    df["MA5"]         = df["Close"].rolling(5).mean()
    df["MA20"]        = df["Close"].rolling(20).mean()
    df["MA_Ratio"]    = df["MA5"] / df["MA20"]
    df.drop(columns=["MA5", "MA20"], inplace=True)

    # ── Log Volume: compresses heavy right tail ───────────────────────────
    df["Log_Volume"]  = np.log(df["Volume"])

    df.replace([np.inf, -np.inf], np.nan, inplace=True)
    df.dropna(inplace=True)
    return df

df = add_features(df)

# Plot engineered features
plot_feats = ["Return","Log_Return","HL_Ratio","Vol_10","MA_Ratio","Log_Volume"]
fig, axes = plt.subplots(3, 2, figsize=(14, 10))
for ax, feat in zip(axes.flatten(), plot_feats):
    ax.plot(df[feat], linewidth=0.7)
    ax.set_title(feat)
    ax.tick_params(labelsize=8)
plt.suptitle("Engineered Features Over Time")
plt.tight_layout()
plt.savefig(r"outputs/plots/03_engineered_features.png", dpi=150)
plt.show()

print(f"Shape after engineering : {df.shape}")
print(f"Columns : {list(df.columns)}")
```

```python
TARGET_COL   = "Log_Return"    # stationary, bounded — predict CHANGES not levels
SEQUENCE_LEN = 60              # 60 trading days ≈ 3 months look-back

FEATURE_COLS = [
    "Open", "High", "Low", "Close", "Volume",
    "Return", "Log_Return",
    "HL_Ratio", "OC_Ratio",
    "Momentum_5", "Momentum_10",
    "Vol_10", "MA_Ratio", "Log_Volume"
]

N_FEATURES = len(FEATURE_COLS)
print(f"Features : {N_FEATURES}")
print(f"Target   : {TARGET_COL}")
```

```python
# Fixed 70/15/15 split:
#   Train : 1996 → 2017  (dot-com, 2008, recovery = 4 regimes)
#   Val   : 2017 → 2021  (COVID crash + V-recovery)
#   Test  : 2021 → 2026  (steady uptrend)

# Problem:
#   Training data had 4 different regimes — very hard to fit
#   Validation had 1 simple trend — very easy
#   → Train Loss 0.69 >> Val Loss 0.22 (gap never closed)
#   → Model never saw a steady trend during training → underfit

# Walk-forward fix:
#   Each fold trains on progressively more data
#   Each fold's validation is immediately AFTER its training period
#   Later folds include the COVID period in TRAINING
#   → Model is exposed to all regime types during training
#   → Train Loss and Val Loss should converge
```

```python
def walk_forward_splits(df, n_folds=5, val_frac=0.15, seq_len=60):
    """
    Creates n_folds walk-forward splits.

    Visual layout (5 folds):
    Fold 1: [===TRAIN===][VAL]
    Fold 2: [=====TRAIN=====][VAL]
    Fold 3: [=======TRAIN=======][VAL]
    Fold 4: [=========TRAIN=========][VAL]
    Fold 5: [===========TRAIN===========][VAL] ← TEST (held out)

    The LAST fold's val set is the final unseen test set.
    All other folds are used for training + validation.
    """
    n       = len(df)
    # Reserve final 15% as permanent held-out test set
    test_start = int(n * (1 - val_frac))
    test_df    = df.iloc[test_start - seq_len :]

    # Remaining data is split into walk-forward folds
    data_df    = df.iloc[:test_start]
    n_data     = len(data_df)

    fold_size  = n_data // n_folds
    splits     = []

    for fold in range(1, n_folds + 1):
        train_end = fold * fold_size
        val_end   = min(train_end + int(n_data * val_frac), n_data)

        train = data_df.iloc[:train_end]
        val   = data_df.iloc[train_end - seq_len : val_end]

        if len(val) < seq_len + 10:
            continue

        splits.append((fold, train, val))
        print(f"  Fold {fold}: Train {train.index[0].date()} → "
              f"{train.index[-1].date()} ({len(train):,} rows) | "
              f"Val {val.index[seq_len].date()} → "
              f"{val.index[-1].date()} ({len(val)-seq_len:,} rows)")

    print(f"\n  Test  : {test_df.index[seq_len].date()} → "
          f"{test_df.index[-1].date()} ({len(test_df)-seq_len:,} rows)")
    return splits, test_df


splits, test_df = walk_forward_splits(df, n_folds=5,
                                      val_frac=0.15,
                                      seq_len=SEQUENCE_LEN)
```

```python
def make_sequences(X: np.ndarray, y: np.ndarray, seq_len: int):
    """
    Sliding window. Overlap rows (first seq_len rows) are INPUT context only.
    They never appear as prediction targets.
    """
    Xs, ys = [], []
    for i in range(seq_len, len(X)):
        Xs.append(X[i - seq_len : i])
        ys.append(y[i])
    return np.array(Xs), np.array(ys)


def preprocess_split(train_df, val_df, feature_cols, target_col, seq_len):
    """
    Fit StandardScaler on train only → transform val.
    Returns sequences + scalers.
    """
    f_scaler = StandardScaler()
    t_scaler = StandardScaler()

    X_tr_raw = f_scaler.fit_transform(train_df[feature_cols])
    X_va_raw = f_scaler.transform(val_df[feature_cols])

    y_tr_raw = t_scaler.fit_transform(train_df[[target_col]])
    y_va_raw = t_scaler.transform(val_df[[target_col]])

    X_tr, y_tr = make_sequences(X_tr_raw, y_tr_raw, seq_len)
    X_va, y_va = make_sequences(X_va_raw, y_va_raw, seq_len)

    return X_tr, y_tr, X_va, y_va, f_scaler, t_scaler
```

```python
class BahdanauAttention(layers.Layer):
    """
    Additive (Bahdanau) attention.

    For all T hidden states from GRU:
        score_i  = tanh(W · h_i)          ← learned scoring
        alpha_i  = softmax(V · score_i)   ← attention weights (sum to 1)
        context  = Σ alpha_i · h_i        ← weighted summary vector

    Returns context vector + alpha weights (for visualization).
    """
    def __init__(self, units, **kwargs):
        super().__init__(**kwargs)
        self.units = units
        self.W     = layers.Dense(units)
        self.V     = layers.Dense(1)

    def call(self, hidden_states):
        score   = tf.nn.tanh(self.W(hidden_states))       # (batch, T, units)
        alpha   = tf.nn.softmax(self.V(score), axis=1)    # (batch, T, 1)
        context = tf.reduce_sum(alpha * hidden_states, axis=1)  # (batch, units)
        return context, alpha

    def get_config(self):
        cfg = super().get_config()
        cfg.update({"units": self.units})
        return cfg


def build_model(seq_len:     int,
                n_features:  int,
                gru1:        int   = 256,
                gru2:        int   = 128,
                attn_units:  int   = 64,
                dense1:      int   = 64,
                dense2:      int   = 32,
                dropout:     float = 0.1,
                l2:          float = 1e-4) -> Model:
    """
    Architecture:
        Input → GRU-256 → BN → GRU-128 → BN → Attention-64
               → Dense-64 → Dropout → Dense-32 → Dense-1

    Key decisions:
        dropout=0.1   : was 0.2 in v2/v3, artificially raised train loss
        L2 on weights : smooth regularization without masking neurons
        return_sequences=True on both GRUs → attention sees ALL timesteps
        clipnorm=1.0  : prevents exploding gradients in GRU-256
    """
    reg    = regularizers.l2(l2)
    inputs = tf.keras.Input(shape=(seq_len, n_features), name="input")

    # ── GRU Layer 1 ──────────────────────────────────────────────────────
    x = layers.GRU(gru1,
                   return_sequences      = True,
                   dropout               = dropout,
                   recurrent_dropout     = 0.0,
                   recurrent_regularizer = reg,
                   name                  = "gru_1")(inputs)
    x = layers.BatchNormalization()(x)

    # ── GRU Layer 2 ──────────────────────────────────────────────────────
    x = layers.GRU(gru2,
                   return_sequences      = True,
                   dropout               = dropout,
                   recurrent_regularizer = reg,
                   name                  = "gru_2")(x)
    x = layers.BatchNormalization()(x)

    # ── Attention ─────────────────────────────────────────────────────────
    context, _ = BahdanauAttention(attn_units, name="attention")(x)

    # ── Regression head ───────────────────────────────────────────────────
    x      = layers.Dense(dense1, activation="relu",
                          kernel_regularizer=reg)(context)
    x      = layers.Dropout(dropout)(x)
    x      = layers.Dense(dense2, activation="relu",
                          kernel_regularizer=reg)(x)
    output = layers.Dense(1, name="log_return_pred")(x)

    return Model(inputs, output, name="AttnGRU_v4")


# Quick architecture check
_m = build_model(SEQUENCE_LEN, N_FEATURES)
_m.summary()
print(f"\nTotal parameters : {_m.count_params():,}")
del _m
```

```python
WARMUP_EPOCHS = 10
INIT_LR       = 1e-4
PEAK_LR       = 1e-3

def warmup_schedule(epoch, lr):
    """Ramp LR from INIT_LR to PEAK_LR over first WARMUP_EPOCHS epochs."""
    if epoch < WARMUP_EPOCHS:
        return INIT_LR + (PEAK_LR - INIT_LR) * (epoch / WARMUP_EPOCHS)
    return lr


def train_fold(fold_num, train_df, val_df):
    """Train one walk-forward fold. Returns trained model + history."""
    print(f"\n{'='*60}")
    print(f"  FOLD {fold_num}  |  "
          f"Train: {len(train_df):,} rows  |  "
          f"Val: {len(val_df)-SEQUENCE_LEN:,} rows")
    print(f"{'='*60}")

    X_tr, y_tr, X_va, y_va, f_sc, t_sc = preprocess_split(
        train_df, val_df, FEATURE_COLS, TARGET_COL, SEQUENCE_LEN
    )

    model = build_model(SEQUENCE_LEN, N_FEATURES)
    model.compile(
        optimizer = tf.keras.optimizers.Adam(
                        learning_rate = INIT_LR,
                        clipnorm      = 1.0       # prevent exploding gradients
                    ),
        loss    = "huber",
        metrics = ["mae"]
    )

    ckpt_path = rf"outputs/models/fold_{fold_num}_best.keras"
    callbacks = [
        EarlyStopping(monitor="val_loss", patience=25,
                      restore_best_weights=True, verbose=0),
        ReduceLROnPlateau(monitor="val_loss", factor=0.5,
                          patience=15, min_lr=1e-7, verbose=0),
        ModelCheckpoint(ckpt_path, monitor="val_loss",
                        save_best_only=True, verbose=0),
        LearningRateScheduler(warmup_schedule, verbose=0)
    ]

    history = model.fit(
        X_tr, y_tr,
        validation_data = (X_va, y_va),
        epochs          = 100,
        batch_size      = 32,
        callbacks       = callbacks,
        verbose         = 1
    )

    best_epoch = int(np.argmin(history.history["val_loss"])) + 1
    best_val   = min(history.history["val_loss"])
    print(f"\n  Best epoch : {best_epoch}  |  Best val_loss : {best_val:.5f}")

    return model, history, f_sc, t_sc


# ── Train all folds ───────────────────────────────────────────────────────
fold_results = []
for fold_num, train_df_fold, val_df_fold in splits:
    model, history, f_sc, t_sc = train_fold(fold_num, train_df_fold, val_df_fold)
    fold_results.append({
        "fold"    : fold_num,
        "model"   : model,
        "history" : history,
        "f_scaler": f_sc,
        "t_scaler": t_sc,
        "train_df": train_df_fold,
        "val_df"  : val_df_fold,
    })

# ── Use the LAST fold's model + scalers for final evaluation ─────────────
# Last fold trained on the most data (best generalization)
best_result   = fold_results[-1]
final_model   = best_result["model"]
final_f_sc    = best_result["f_scaler"]
final_t_sc    = best_result["t_scaler"]
final_train   = best_result["train_df"]

print("\nFinal model is from fold", best_result["fold"])
```

```python
# ── LR MAPE REMOVED ──────────────────────────────────────────────────────
# Log Return MAPE is mathematically misleading:
#   typical log return = 0.001
#   tiny absolute error = 0.002 → MAPE = 200%  (looks terrible but is fine)
# All evaluation is done in ₹ price space using rolling anchor reconstruction

def reconstruct_price_rolling(lr_true: np.ndarray,
                               lr_pred: np.ndarray,
                               anchor_prices: np.ndarray,
                               reset_every: int = 20) -> tuple:
    """
    Reconstruct ₹ price using a rolling anchor reset every N days.

    WHY: single cumsum over 1,000 days → early errors compound to all later
    predictions. In v2 this caused val MAPE=43% even with good per-day
    predictions. Reset every 20 days caps compounding to at most 20 steps.

    anchor_prices: the TRUE Close price at each prediction date.
                   Used only as the reset anchor — not as a target.
    """
    prices_true = np.zeros(len(lr_true))
    prices_pred = np.zeros(len(lr_true))

    for i in range(len(lr_true)):
        if i % reset_every == 0:
            anchor = anchor_prices[i]

        start         = i - (i % reset_every)
        prices_true[i] = anchor * np.exp(np.sum(lr_true[start : i+1]))
        prices_pred[i] = anchor * np.exp(np.sum(lr_pred[start : i+1]))

    return prices_true, prices_pred


def evaluate(model, X, y_scaled, t_scaler,
             anchor_series: pd.Series,
             label: str,
             reset_every: int = 20) -> tuple:
    """
    Evaluate model in ₹ price space.
    Uses rolling anchor reconstruction — no cumsum compounding.

    Returns (y_true_price, y_pred_price) for plotting.
    """
    y_pred_s  = model.predict(X, verbose=0)
    y_pred_lr = t_scaler.inverse_transform(y_pred_s).flatten()
    y_true_lr = t_scaler.inverse_transform(y_scaled).flatten()

    anchor_prices           = anchor_series.values
    y_true_price, y_pred_price = reconstruct_price_rolling(
        y_true_lr, y_pred_lr, anchor_prices, reset_every
    )

    mae  = mean_absolute_error(y_true_price, y_pred_price)
    rmse = np.sqrt(mean_squared_error(y_true_price, y_pred_price))
    mape = np.mean(np.abs((y_true_price - y_pred_price) / y_true_price)) * 100
    r2   = r2_score(y_true_price, y_pred_price)

    print(f"\n── {label}  [rolling anchor reset={reset_every}d] ────────────────")
    print(f"  MAE   : ₹{mae:.2f}")
    print(f"  RMSE  : ₹{rmse:.2f}")
    print(f"  MAPE  : {mape:.3f} %")
    print(f"  R²    : {r2:.4f}")

    return y_true_price, y_pred_price
```

```python
def forecast_future(model, last_sequence: np.ndarray,
                    t_scaler, last_known_price: float,
                    n_days: int = 30) -> list:
    """
    Autoregressively predicts n_days beyond last known date.
    Operates in log-return space → converts to ₹ at the end.
    """
    seq           = last_sequence.copy()
    log_ret_preds = []
    lr_idx        = FEATURE_COLS.index("Log_Return")

    for _ in range(n_days):
        pred_s  = model.predict(seq, verbose=0)
        pred_lr = t_scaler.inverse_transform(pred_s)[0, 0]
        log_ret_preds.append(pred_lr)

        new_row         = seq[0, -1, :].copy()
        new_row[lr_idx] = pred_s[0, 0]
        seq             = np.roll(seq, -1, axis=1)
        seq[0, -1, :]   = new_row

    return (last_known_price * np.exp(np.cumsum(log_ret_preds))).tolist()


last_known_price = float(test_df["Close"].iloc[-1])
future_preds     = forecast_future(final_model, X_test[-1:],
                                   final_t_sc, last_known_price, n_days=30)
future_dates     = pd.bdate_range(
    start   = test_dates[-1] + pd.Timedelta(days=1),
    periods = 30
)

plt.figure(figsize=(15, 5))
plt.plot(test_dates[-90:], y_test_true[-90:],
         label="Historical Actual", color="#2196F3", linewidth=1.5)
plt.plot(test_dates[-90:], y_test_pred[-90:],
         label="Model (test fit)",  color="#FF5722",
         linewidth=1.2, linestyle="--")
plt.plot(future_dates, future_preds,
         label="30-Day Forecast",   color="#4CAF50",
         linewidth=2, marker="o", markersize=4)
plt.axvline(test_dates[-1], color="black", linestyle=":",
            linewidth=1.2, label="Forecast Start")
plt.title(f"30-Day Future Forecast  (anchor = ₹{last_known_price:.0f})")
plt.xlabel("Date")
plt.ylabel("₹ Price")
plt.legend()
plt.tight_layout()
plt.savefig(r"outputs/plots/11_future_forecast_30days.png", dpi=150)
plt.show()

pd.DataFrame({
    "Date"            : future_dates,
    "Predicted_Close" : future_preds
}).to_csv(r"outputs/results/future_forecast_30days.csv", index=False)
print("Future forecast saved.")
```

```python
# ── Per-fold validation MAPE ─────────────────────────────────────────────
fold_metrics = []
for res in fold_results:
    fval_df  = res["val_df"]
    fmodel   = res["model"]
    ft_sc    = res["t_scaler"]
    ff_sc    = res["f_scaler"]

    Xv_raw   = ff_sc.transform(fval_df[FEATURE_COLS])
    yv_raw   = ft_sc.transform(fval_df[[TARGET_COL]])
    Xv, yv   = make_sequences(Xv_raw, yv_raw, SEQUENCE_LEN)
    anc      = fval_df["Close"].iloc[SEQUENCE_LEN:].reset_index(drop=True)

    yt, yp   = evaluate(fmodel, Xv, yv, ft_sc, anc,
                        f"Fold {res['fold']} Val", reset_every=20)
    fold_metrics.append({
        "Fold"  : res["fold"],
        "MAPE %": round(np.mean(np.abs((yt-yp)/yt))*100, 3),
        "R²"    : round(r2_score(yt, yp), 4),
    })

# ── Test set final metrics ────────────────────────────────────────────────
test_mape = round(np.mean(np.abs((y_test_true-y_test_pred)/y_test_true))*100, 3)
test_r2   = round(r2_score(y_test_true, y_test_pred), 4)
test_mae  = round(mean_absolute_error(y_test_true, y_test_pred), 2)
test_rmse = round(np.sqrt(mean_squared_error(y_test_true, y_test_pred)), 2)

print("\n── Per-Fold Validation Metrics ─────────────────────────────────")
print(pd.DataFrame(fold_metrics).to_string(index=False))

print("\n── Final Test Set Metrics ──────────────────────────────────────")
print(f"  MAE   : ₹{test_mae}")
print(f"  RMSE  : ₹{test_rmse}")
print(f"  MAPE  : {test_mape} %")
print(f"  R²    : {test_r2}")

# ── Save summary CSV ──────────────────────────────────────────────────────
summary_rows = fold_metrics + [{
    "Fold": "TEST", "MAPE %": test_mape, "R²": test_r2
}]
pd.DataFrame(summary_rows).to_csv(
    r"outputs/results/error_metrics.csv", index=False
)

# ── MAPE bar chart ────────────────────────────────────────────────────────
labels = [f"Fold {r['Fold']}" for r in fold_metrics] + ["TEST"]
mapes  = [r["MAPE %"] for r in fold_metrics] + [test_mape]
colors = ["#42A5F5"] * len(fold_metrics) + ["#EF5350"]

plt.figure(figsize=(10, 4))
bars = plt.bar(labels, mapes, color=colors)
for bar, val in zip(bars, mapes):
    plt.text(bar.get_x() + bar.get_width()/2,
             bar.get_height() + 0.05,
             f"{val:.2f}%", ha="center", fontsize=9)
plt.title("MAPE % — Per Fold Validation + Final Test")
plt.ylabel("MAPE %")
plt.tight_layout()
plt.savefig(r"outputs/plots/12_mape_all_folds.png", dpi=150)
plt.show()
```
